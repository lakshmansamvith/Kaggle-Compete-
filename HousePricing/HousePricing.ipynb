# Checking for Null Values 

```
null_values = train_set.isnull().sum()
test_null_values = train_set.isnull().sum()
drop_columns = [(null_values.index[count]) for count, val in enumerate(null_values) if val/len(train_set)>0.5]
test_drop_columns = [(null_values.index[count]) for count, val in enumerate(null_values) if val/len(test_set)>0.5]
print(drop_columns, len(train_set.columns))
print(test_drop_columns)
```

# Getting information on numerical and non-numerical columns for imputing values

```
numerical_values = [(col) for ind, col in enumerate(train_set.columns[1:76]) if (isinstance(train_set[col][ind],  np.int64) or 
                                                                               isinstance(train_set[col][ind],  np.float64))]
numerical_values
non_numerical_values = [(col) for ind, col in enumerate(train_set.columns[1:76]) if col not in numerical_values]
numerical_values
```

#Separating Input and Target Columns

```
X_df = train_set.copy()
```

#Imputing NaN values

```
from sklearn.impute import SimpleImputer 
imputer = SimpleImputer(missing_values=np.NaN, strategy='mean')
imputer.fit(X_df[numerical_values])
X_df[numerical_values] = imputer.transform(X_df[numerical_values])
test_set[numerical_values] = imputer.transform(test_set[numerical_values])
```

```
sums = test_set.isnull().sum()
for ind, val in enumerate(sums):
        if (isinstance(X_df[sums.index[ind]][1], np.int64) or isinstance(X_df[sums.index[ind]][1], np.float64)):
            print(sums.index[ind], val, type(X_df[sums.index[ind]][1]))
```

# Ordinal Encoding

```
from sklearn.preprocessing import OrdinalEncoder

X_df[non_numerical_values] = X_df[non_numerical_values].replace(np.nan, '?', inplace=True)
test_set[non_numerical_values] = test_set[non_numerical_values].replace(np.nan, '?', inplace=True)
encoder = OrdinalEncoder()
encoder.fit(X_df[non_numerical_values])
X_df[non_numerical_values] = encoder.transform(X_df[non_numerical_values])
test_set[non_numerical_values] = encoder.transform(test_set[non_numerical_values])
```

#Imputing Categorical Values

```
categorical_imputer = SimpleImputer(missing_values=np.NaN, strategy='most_frequent')
categorical_imputer.fit(X_df[non_numerical_values])
X_df[non_numerical_values] = categorical_imputer.transform(X_df[non_numerical_values])
test_set[non_numerical_values] = categorical_imputer.transform(test_set[non_numerical_values])

```
